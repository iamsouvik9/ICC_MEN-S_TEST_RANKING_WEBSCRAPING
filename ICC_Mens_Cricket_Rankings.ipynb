```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://www.icc-cricket.com/rankings/mens/player-rankings/test/batting"
page=requests.get(url).text
soup=BeautifulSoup(page,'html.parser')
```

```python
soup.prettify()
```

```python
print('Classes of each table:')
for table in soup.find_all('table'):
    print(table.get('class'))
```

```python
table = soup.find('table', class_='table rankings-table')
```

```python
for ele in soup.find_all('tr'):
  print(ele)
  print("\n")    

```

```python
position=[]



for ele in soup.find_all('span',class_="rankings-block__pos-number"):
  position.append(ele.text.strip())

for ele in soup.find_all('span',class_="rankings-table__pos-number"):
  position.append(ele.text.strip())
```

```python
player=[]
lst=[]
for ele in soup.find_all('a'):
  lst.append(ele.text.strip())
player=lst[87:187]
```

```python
team=[]
team.append('ENG')
for ele in soup.find_all('span',class_="table-body__logo-text"):
  team.append(ele.text.strip())
```

```python
ratings=[]
ratings.append(soup.find('div',class_="rankings-block__banner--rating").text.strip())
for ele in soup.find_all('td',class_="table-body__cell rating"):
  ratings.append(ele.text.strip())
len(ratings)
```

```python
dict1={"SLN":position,"NAME":player,"COUNTRY":team,"RATINGS":ratings}
```

```python
df=pd.DataFrame.from_dict(dict1)
df.head()
```
